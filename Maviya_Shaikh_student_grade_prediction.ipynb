## Student Grade Prediction
The analysis of two student datasets was conducted to forecast their final grade. The two datasets are based on two different subjects Mathematics and Portugese language. There are total 33 features in both the dataset from which 32 are independent variables and 1 is dependent variable i.e. final grade(G3) that is predicted in this project.

Independent variable:
1. school - student's school (binary: "GP" - Gabriel Pereira or "MS" - Mousinho da Silveira)
2. sex - student's sex (binary: "F" - female or "M" - male)
3. age - student's age (numeric: from 15 to 22)
4. address - student's home address type (binary: "U" - urban or "R" - rural)
5. famsize - family size (binary: "LE3" - less or equal to 3 or "GT3" - greater than 3)
6. Pstatus - parent's cohabitation status (binary: "T" - living together or "A" - apart)
7. Medu - mother's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
8. Fedu - father's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
9. Mjob - mother's job (nominal: "teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other")
10. Fjob - father's job (nominal: "teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other")
11. reason - reason to choose this school (nominal: close to "home", school "reputation", "course" preference or "other")
12. guardian - student's guardian (nominal: "mother", "father" or "other")
13. traveltime - home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)
14. studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
15. failures - number of past class failures (numeric: n if 1<=n<3, else 4)
16. schoolsup - extra educational support (binary: yes or no)
17. famsup - family educational support (binary: yes or no)
18. paid - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)
19. activities - extra-curricular activities (binary: yes or no)
20. nursery - attended nursery school (binary: yes or no)
21. higher - wants to take higher education (binary: yes or no)
22. internet - Internet access at home (binary: yes or no)
23. romantic - with a romantic relationship (binary: yes or no)
24. famrel - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
25. freetime - free time after school (numeric: from 1 - very low to 5 - very high)
26. goout - going out with friends (numeric: from 1 - very low to 5 - very high)
27. Dalc - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
28. Walc - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)
29. health - current health status (numeric: from 1 - very bad to 5 - very good)
30. absences - number of school absences (numeric: from 0 to 93)
31. G1 - first period grade (numeric: from 0 to 20)
32. G2 - second period grade (numeric: from 0 to 20)

Dependent Variable (target variable):
33. G3 - final grade (numeric: from 0 to 20, output target)

```python
#Checking correlation between variables
def correlation(dataset,threshold):
    count=0
    col_corr= [] # List of correlated columns
    corr_matrix=dataset.corr() #finding correlation between columns
    for i in range (len(corr_matrix.columns)): #Number of columns
        for j in range (i):
            if abs(corr_matrix.iloc[i,j])>threshold: #checking correlation between columns
                colName=(corr_matrix.columns[i], corr_matrix.columns[j]) #getting correlated columns
                count=count+1
                col_corr.append(colName) #adding correlated column name
    print("Total num of correlation pairs: ",count)
    return col_corr #returning set of column names
    
col=correlation(mat_df,0.80)
print('Correlated columns for Maths dataset @ 0.80:', col)
print("-----------------------------------------------------")
ol=correlation(por_df,0.80)
print('Correlated columns for Portugese dataset @ 0.80:', col)
```

# ------------------------------------Development Stage-----------------------------------------

80% of the data is used for training the regression model and 20% of the data is used for testing the model in all the development stages.

```python
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics  
```

## Development Stage 1:- using original dataset with all the variables

```python
#Define x and y variable
x = mat_df.drop('G3',axis=1).values
y = mat_df['G3'].values
```

```python
#dividing the dataset into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
#Script for Linear Regression
for name,method in [('Linear Regression', LinearRegression())]: 
    method.fit(x_train,y_train)
    predict = method.predict(x_test)

print('Method: {}'.format(name))   

#MAE,MSE and RMSE
print('\nR2: {:0.2f}'.format(metrics.r2_score(y_test, predict)))
print('Mean Absolute Error: {:0.2f}'.format(metrics.mean_absolute_error(y_test, predict)))  
print('Mean Squared Error: {:0.2f}'.format(metrics.mean_squared_error(y_test, predict)))  
print('Root Mean Squared Error: {:0.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, predict))))
```

```python
#Define x and y variable
x = por_df.drop('G3',axis=1).values
y = por_df['G3'].values


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)

for name,method in [('Linear Regression', LinearRegression())]: 
    method.fit(x_train,y_train)
    predict = method.predict(x_test)

print('Method: {}'.format(name))   
    
#MAE,MSE and RMSE
print('\nR2: {:0.2f}'.format(metrics.r2_score(y_test, predict)))
print('Mean Absolute Error: {:0.2f}'.format(metrics.mean_absolute_error(y_test, predict)))  
print('Mean Squared Error: {:0.2f}'.format(metrics.mean_squared_error(y_test, predict)))  
print('Root Mean Squared Error: {:0.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, predict))))
```

##### For both the dataset, R2 of 0.83 indicates strong relationship between independent and dependent variable.Also, the errors in both the dataset are similar and not that much high. So, both these models in stage 1 are good models.

## Development Stage 3 - using original dataset with Feature Selection

We have lot of features in the dataset and as we have check the correlation between variables, we have found out that only two variables has more than 80% correlation with dependent variable G3. So in stage 3 and stage 4 we will do feature selection i.e.only G1 and G2 is used for the prediction of G3.

```python
#Define x and y variable
x = mat_df[['G1','G2']].values
y = mat_df['G3'].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)


for name,method in [('Linear Regression', LinearRegression())]: 
    method.fit(x_train,y_train)
    predict = method.predict(x_test)

print('Method: {}'.format(name))   

#Coefficents
print('\nIntercept: {:0.2f}'.format(float(method.intercept_)))
coeff_table=pd.DataFrame(np.transpose(method.coef_),mat_df[['G1','G2']].columns,columns=['Coefficients'])
print('\n')
print(coeff_table)
    
#MAE,MSE and RMSE
print('\nR2: {:0.2f}'.format(metrics.r2_score(y_test, predict)))
print('Mean Absolute Error: {:0.2f}'.format(metrics.mean_absolute_error(y_test, predict)))  
print('Mean Squared Error: {:0.2f}'.format(metrics.mean_squared_error(y_test, predict)))  
print('Root Mean Squared Error: {:0.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, predict))))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)


#Script for Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn import metrics  

for name,method in [('Linear Regression', LinearRegression())]: 
    method.fit(x_train,y_train)
    predict = method.predict(x_test)

print('Method: {}'.format(name))   

#Coefficents
print('\nIntercept: {:0.2f}'.format(float(method.intercept_)))
coeff_table=pd.DataFrame(np.transpose(method.coef_),new_mat_df[['G1','G2']].columns,columns=['Coefficients'])
print('\n')
print(coeff_table)
    
#MAE,MSE and RMSE
print('\nR2: {:0.2f}'.format(metrics.r2_score(y_test, predict)))
print('Mean Absolute Error: {:0.2f}'.format(metrics.mean_absolute_error(y_test, predict)))  
print('Mean Squared Error: {:0.2f}'.format(metrics.mean_squared_error(y_test, predict)))  
print('Root Mean Squared Error: {:0.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, predict))))
```
